````python
freq_min = 27
freq_max = 1700
# Definindo Bandas
band_max = [77, 1766]
band_VHF = [30, 300]
band_UHF = [300,1000]
band_L = [1000, 1766]

# Rodar os shell comandos de uma maneira intuitiva usando shlex e pegando o output para a tela.
def run_command(command, out = True):
    process = subprocess.Popen(shlex.split(command), stdout=subprocess.PIPE)
    out_list = []
    while True:
        output = process.stdout.readline().decode()
        out_list.append(output)
        if output == '' and process.poll() is not None:
            break
        if output:
            print(output.strip())
    if out == True:
        result = process.poll()
    else:
        result = out_list
    return result

# Sincroniza arquivos entre raspberry e local.
def sinc_files(local_folder = "../data/external/SDR", remote_folder = "~/SDR"):
    command =  "scp -r pi@" + IP + ":" + remote_folder + " " + local_folder
    result = run_command(command);
    return None

# Sincroniza arquivos entre raspberry e local.
def clean(mask = "~/SDR/*.csv"):
    command =  "ssh pi@" + IP + " rm " + mask
    result = run_command(command);
    return None


# Plota spectro contido em DF na banda de frequência indicada.
def plot_spectrum(df = None, band = None, freq_min = freq_min, freq_max = freq_max, ax = None):
    if ax is not None:
        fig, ax = plt.subplots(figsize = (18,6))
    if isinstance(band, list) is None:
        band = [freq_min, freq_max]
    ax = df.plot("Freq", "Flux", ax = ax)
    ax.set_ylabel("Fluxo (dB)")
    ax.set_xlabel("Frequência (MHz)")
    ax.set_xlim(band)
    ax.legend()
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.set_axisbelow(True)
    ax.grid(color = "lightgray", linewidth = 1)
    return ax
````

## Planejando uma observação no Labmet

### Preparando observação

````python
# Definindo Instrumento
name_device = "BINGO"
path = path="../data/raw/"
MiniHorn = Uirapuru.set_observatory()
MiniHorn.fwhm = 45
name_device = "BINGO"
backend = RTLSDR_power(name=name_device, path=path)
backend.modes = {"SKY":'01'}
MiniHorn.backend = backend
# Criando observação
begin = pd.to_datetime("2022-02-28 00:00:00")
#trial_run = radiotools.Observations().load_daily_data(begin = begin, instrument = Uirapuru)
````

````python
# Definindo apontamento e pegando coordenadas de objetos
begin = MiniHorn.timezone.localize(pd.to_datetime("2022-03-27 10:00:00"))
end = MiniHorn.timezone.localize(pd.to_datetime("2022-03-27 17:00:00"))
duration = (end-begin).total_seconds()
duration_obs = duration*u.s + 900*u.s

trial_run = Observations(t_start = begin,
                         duration = duration_obs, 
                         instrument = MiniHorn).make_timevector(delta = 1 * u.min, 
                                                                inplace = True).make_pointings()
df_sky = trial_run.get_all_beam()
df_sky["TIME"] = pd.to_datetime(df_sky.TIME.values, unit = "D", origin = "julian")
````

````python
# Preparando informações ALTAZ
df_sky_time = trial_run.get_all_beam()
df_altaz = trial_run.get_altaz_from_radec(objects = df_sky_time)
df_altaz["TIME"] = pd.to_datetime(df_altaz.TIME.values, unit = "D", origin = "julian", utc = True)

time_vec = df_altaz.TIME.unique()
time_len = time_vec.shape[0]
time_markers = time_vec[np.arange(0,time_len, np.ceil(time_len / 10)).astype(int)]
df_altaz = df_altaz[df_altaz.TIME.isin(time_markers)]
````

## Planejamento de Medida

````python
def scan(prefix = None, 
         local_folder = None, 
         band = None, bandwidth = 1, 
         integration = "1s", 
         elapsed = "15m", 
         filename = None, 
         local_folder = None
        ):
    NOW = datetime.now().strftime("%d_%m_%Y_%H_%M_%S")
    FILENAME = prefix + NOW + ".csv"
    FILENAME_remote = "~/SDR/" + FILENAME
    scan_band = ( command + "rtl_power -f " + str(band[0]) + "M:" + str(band[1]) + "M:" + 
                str(bandwidth) + "M -g 50" +  " -i " + integration + " -e " + elapsed + " " + FILENAME_remote)
    print(scan)
    print("conectando")
    run_command(scan);
    print("sincronizando")
    sinc_files(local_folder=local_folder)
    print("cleaning")
    clean(mask = FILENAME_remote)
    print("done!")
    return None
````
