# Arquivo FITS

```python
import numpy as np
import astropy
from astropy.io import fits
from astropy.table import Table
from datetime import datetime
```

```python
NOW = datetime.now()
teste = datetime.strftime(NOW, "%Y")
DATA_START_string = TIME_START_string = TIME_END_string = DATA_END_string = teste
```

```python
header = fits.Header()
header["SIMPLE"] = "T"
header["BITPIX"] = "16"
header["NAXIS"] = 2
header["NAXIS1"] = time_size
header["NAXIS2"] = freq_size
header["EXTEND"] = "T"
header["COMMENT"] = "FITS (Flexible Image Transport System) format defined in Astronomy and"
header["COMMENT"] = "Astrophysics Supplement Series v44#p363, v44#p371, v73#p359, v73#p365."
header["COMMENT"] = "Contact the NASA Science Office of Standards and Technology for the   "
header["COMMENT"] = "FITS Definition document #100 and other FITS information.             "
header["DATE"] = '2021-09-01'         # Time of observation                            
header["CONTENT"] = 'Radio flux density, RTL_SDR (BINGO)' # Title of image 
header["ORIGIN"] = 'PARAIBA'           # Organization name                              
header["TELESCOP"] = 'MiniHorn' # Type of instrument                             
header["INSTRUME"] = 'RTLSDR'           # Name of the spectrometer                       
header["OBJECT"] = 'Sky'           # object description                             
header["DATE-OBS"] = DATA_START_string         #Date observation starts                        
header["TIME-OBS"] = TIME_START_string       # Time observation starts                        
header["DATE-END"] = DATA_END_string         # date observation ends                          
header["TIME-END"] = TIME_END_string           # time observation ends                          
header["BZERO"] = 0. # scaling offset                                 
header["BSCALE"] = 1. # scaling factor                                 
header["BUNIT"] = 'digits'           # z-axis title                                   
header["DATAMIN"] = 0 # minimum element in image                       
header["DATAMAX"] = 254 # maximum element in image                       
header["CRVAL1"] =  541. # value on axis 1 at reference pixel [sec of day]
header["CRPIX1"] =  0 # reference pixel of axis 1                      
header["CTYPE1"] = 'Time [UT]'          # title of axis 1                                
header["CDELT1"] =  0.5 # step between first and second element in x-axis
header["CRVAL2"] =  400. # value on axis 2 at the reference pixel         
header["CRPIX2"] =  0 # reference pixel of axis 2                      
header["CTYPE2"] = 'Frequency [MHz]'    # title of axis 2                                
header["CDELT2"] =  -1. # step between first and second element in axis  
header["OBS_LAT"] = -7.21181106567383 # observatory latitude in degree                 
header["OBS_LAC"] = 'S       '           # observatory latitude code {N,S}                
header["OBS_LON"] = -35.9081726074219 # observatory longitude in degree                
header["OBS_LOC"] = 'W       '           # observatory longitude code {E,W}               
header["OBS_ALT"] = 17. # observatory altitude in meter asl              
header["FRQFILE"] = 'FRQ00400.CFG'       # name of frequency file                         
header["PWM_VAL"] = 120 # RTLSDR gain setting
header["SDRGAIN"] = 49 # RTLSDR gain setting
header["SDRBW"] = 49 # RTLSDR gain setting
header["SDRFREQS"] = 49 # RTLSDR gain setting
header["SDRTINT"] = 49 # RTLSDR gain setting
header["SDRTELAP"] = 49 # RTLSDR gain setting
```

```python
freq_size = 1024
time_size = 2048

freq_array = np.random.random(size = freq_size)
time_array = np.random.random(size = time_size)

data = np.random.random(size = (time_size,freq_size))

primary_HDU = fits.PrimaryHDU(header = header, data = data )
table_hdu = fits.table_to_hdu(Table([[time_array], [freq_array]], names = ("TIME", "FREQUENCY")))

hdul = fits.HDUList([primary_HDU, table_hdu])
hdul.writeto("./new.fit", overwrite = True)
```
